```python
scale = image[775:780, 800:]

plt.figure(figsize = (10, 1))

ax = plt.subplot(2, 1, 1)
plt.imshow(image[775:], cmap = 'gray')

ax = plt.subplot(2, 1, 2)
plt.imshow(scale, cmap = 'gray')
plt.show()


a = []
for row in scale:
    l = len(row)
    cnt = 0
    flag = False
    for i in range(l - 1):
        #print("i:%d row[i]:%d cnt:%d"%(i, row[i], cnt))
        if row[i] == 255 and row[i + 1] != 255:
            if flag == True:
                flag = False
                a.append(cnt)
                cnt = 0
            else:
                flag = True
        if flag == True:
            cnt += 1
            

print(a)
rule = int(np.mean(a))
print(rule)
    
```

```python
fx = [[0, 1], [1, 0], [0, -1], [-1, 0]]
shape = smooth.shape

def dfs(x, y):
    if x >= shape[0] or x < 0 or y >= shape[1] or y < 0 or smooth[x][y] == 1:
        return 0, 0, 0
    
    ans = 1
    xsum = x
    ysum = y
    smooth[x][y] = 1
    for i in range(4):
        a, b, c = dfs(x + fx[i][0], y + fx[i][1])
        ans += a
        xsum += b
        ysum += c

    return ans, xsum, ysum
      
```

```python
img = deepcopy(smooth)

shape = smooth.shape
print(shape)
print(smooth)
res = []
cnt = 0
for i in range(shape[0]):
    for j in range(shape[1]):
        if smooth[i][j] == 0:
            ans, x, y = dfs(i, j)
            if (ans <= 4): continue
            cnt += 1
            res.append([cnt, x//ans, y//ans, ans])
        
print("It was finished!")

print(res)
```

```python
lower = 30000
upper = 300000
```

```python
res = np.array(res)
plt.figure(figsize = (20, 20))

plt.imshow(image, cmap = 'gray')

for i in res:
    if i[3] >= lower/(rule * rule) and i[3] <= upper/(rule * rule):
        plt.annotate("  " + str(i[0]), xy = (i[2],i[1]), color = 'b', weight = "black", 
                     bbox = dict(boxstyle = 'round,  pad = 0.5', fc = 'y', ec = 'k', lw = 1, alpha = 0.5))
        plt.scatter(i[2], i[1], color = "r", alpha = 1)
        
plt.show()


for i in res:
    if i[3] >= lower/(rule * rule) and i[3] <= upper/(rule * rule):
        print("%5d: (%4d, %4d)  %7d um^2"%(i[0], i[2],i[1], i[3] * rule * rule))
```
